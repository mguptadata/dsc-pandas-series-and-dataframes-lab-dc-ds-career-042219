## Change LINENAME to be the Index

## Painstakingly change the index back

```python
# Painstakingly change the index back
df = df.reset_index() 
df.head()
#### important to note that the order of the columns has changed!
### LINENAME has now been moved/inserted back in as the first column in the dataset
###might want to create a copy so you can keep your original dataset/dataframe in tact
```

```python
df['Num_Lines'] = df.LINENAME.map(lambda x: len(x))

#.map= apply to whole dataset
#new list = map (addition, numbers) -- give me a new list, and the values are equal to all the 
#numbers passing through the function
#format for map is .map(expression, values/list) #because we didn't designate the value,
#it will automatically attach to LINENAME indicated before .map

#tuple is an immutable list (x,y)
```

## Write a function to clean a column name

```python
def clean(col_name):
    cleaned = col_name.strip()  #strips away the whitespace before/after col name 
    return cleaned
```

```python
#This is a list comprehension. It applies your clean function to every item in the list.
#We then reassign that to df.columns
#You shouldn't have to change anything here.
#Your function above should work appropriately here.
df.columns = [clean(col) for col in df.columns] 
```
